### opening a csv file from an URL

```python
import pandas as pd
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

response = requests.get(url)

df = pd.read_csv(StringIO(response.text))

print(df.head())
```
