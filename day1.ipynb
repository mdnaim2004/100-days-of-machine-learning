# AI vs ML vs DL
- Deep Learning is a subset of Machine Learning
- Machine Learning is a subset of Artificial Intelligence
```
--------------------------------------------------------------
|                                                            |
|                 AI - Artificial Intelligence               |
|                                                            |
|     --------------------------------------------------     |
|     |                                                |     |
|     |          ML - Machine Learning                 |     |
|     |                                                |     |
|     |      ----------------------------------        |     |
|     |      |                                |        |     |
|     |      |      DL - Deep Learning        |        |     |
|     |      |                                |        |     |
|     |      ----------------------------------        |     |
|     |                                                |     |
|     --------------------------------------------------     |
|                                                            |
--------------------------------------------------------------
```
---
```
AI (Artificial Intelligence)
|
|--> Making machines intelligent
|--> Broadest field
|--> Example:
|      Chatbots, Robotics, Expert Systems
|
|--> ML (Machine Learning)
|      |
|      |--> Subset of AI
|      |--> Machines learn from data
|      |--> Example:
|             Spam Detection
|             Recommendation System
|
|--> DL (Deep Learning)
       |
       |--> Subset of ML
       |--> Uses Neural Networks with many layers
       |--> Works very well on huge data
       |--> Example:
              ChatGPT
              Self-driving cars
              Face Recognition

```
---
1. AI - Smart behavior

2. ML - Learning from data

3. DL - Learning using deep neural networks

## Types of Machine Learning
---
---
```
Machine Learning
|
|--> Supervised Learning
|        |
|        |--> Uses Labeled Data
|        |--> Input and Output both are given
|        |--> Model learns mapping:
|        |        Input  ---> Output
|        |
|        |--> Example:
|        |       Student Study Hours ---> Exam Marks
|        |
|        |--> Two Main Types:
|               |
|               |--> Regression
|               |       |
|               |       |--> Output is Numerical / Continuous
|               |       |
|               |       |--> Example:
|               |       |       House Size ---> House Price
|               |       |
|               |       |--> Real Example:
|               |       |       Predicting temperature
|               |       |       Predicting salary
|               |       |       Predicting stock price
|               |       |
|               |       |--> Algorithms:
|               |               Linear Regression
|               |               Ridge Regression
|               |               Random Forest Regressor
|               |
|               |
|               |--> Classification
|                       |
|                       |--> Output is Categorical
|                       |
|                       |--> Example:
|                       |       Email ---> Spam or Not Spam
|                       |
|                       |--> Real Example:
|                       |       Cat vs Dog classification
|                       |       Disease Detection
|                       |       Face Mask Detection
|                       |
|                       |--> Algorithms:
|                               Logistic Regression
|                               Decision Tree
|                               SVM
|                               Neural Network
|
|
|--> Unsupervised Learning
|        |
|        |--> Uses Unlabeled Data
|        |--> No Output column is given
|        |--> Model tries to discover hidden patterns
|        |
|        |--> Example:
|        |       Customer data without labels
|        |
|        |--> Types:
|               |
|               |--> Clustering
|               |       |
|               |       |--> Groups similar data points together
|               |       |
|               |       |--> Example:
|               |       |       Grouping customers based on shopping behavior
|               |       |
|               |       |--> Real Example:
|               |       |       Netflix movie recommendation grouping
|               |       |       Friend suggestion system
|               |       |
|               |       |--> Algorithms:
|               |               K-Means
|               |               DBSCAN
|               |               Hierarchical Clustering
|               |
|               |
|               |--> Dimensionality Reduction
|               |       |
|               |       |--> Reduces number of features
|               |       |--> Keeps important information
|               |       |
|               |       |--> Example:
|               |       |       100 features ---> 10 important features
|               |       |
|               |       |--> Real Example:
|               |       |       Image compression
|               |       |       Data visualization
|               |       |
|               |       |--> Algorithms:
|               |               PCA
|               |               t-SNE
|               |
|               |
|               |--> Anomaly Detection
|               |       |
|               |       |--> Detects unusual or abnormal data
|               |       |
|               |       |--> Example:
|               |       |       Fraud Transaction Detection
|               |       |
|               |       |--> Real Example:
|               |       |       Credit Card Fraud Detection
|               |       |       Network Intrusion Detection
|               |       |
|               |       |--> Algorithms:
|               |               Isolation Forest
|               |               One-Class SVM
|               |
|               |
|               |--> Association Rule Learning
|                       |
|                       |--> Finds relationships between items
|                       |
|                       |--> Example:
|                       |       Bread buyers also buy Butter
|                       |
|                       |--> Real Example:
|                       |       Market Basket Analysis
|                       |       Product Recommendation
|                       |
|                       |--> Algorithms:
|                               Apriori
|                               FP-Growth
|
|
|--> Semi-Supervised Learning
|        |
|        |--> Uses both labeled and unlabeled data
|        |
|        |--> Small amount of labeled data
|        |--> Large amount of unlabeled data
|        |
|        |--> Example:
|        |       100 labeled images + 10,000 unlabeled images
|        |
|        |--> Real Example:
|        |       Medical image classification
|        |       Speech recognition
|
|
|--> Reinforcement Learning
         |
         |--> Learning through rewards and punishments
         |--> Agent interacts with environment
         |
         |--> Goal:
         |       Maximize Reward
         |
         |--> Example:
         |       Game Playing AI
         |
         |--> Real Example:
         |       Self-driving car
         |       Chess AI
         |       Robot navigation
         |
         |--> Main Components:
                 Agent
                 Environment
                 Reward
                 Action

```
